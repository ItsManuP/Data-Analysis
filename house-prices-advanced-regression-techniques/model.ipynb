```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn import model_selection
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('train.csv')
```

```python
train.info()
```

```python
train.shape
```

```python
train_subset = train.dropna(axis=1)
```

```python
train_subset.head(10)
```

```python
train_subset = train_subset[["SalePrice","MSSubClass","LotArea","OverallQual","YearBuilt","Foundation","GarageArea","YrSold","MoSold","SaleType","SaleCondition"]]
train_subset.head(5)
```

```python
train_subset_dummies = pd.get_dummies(data= train_subset,drop_first=True, dtype=int)
```

```python
train_subset_dummies.head(1).T
```

```python
y = train_subset_dummies['SalePrice']
X = train_subset_dummies.drop(['SalePrice'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python

ln = LinearRegression()
ln.fit(X_train,y_train)
y_pred = ln.predict(X_test)
```

```python
from sklearn.metrics import r2_score

accuracy = r2_score(y_test,y_pred)
print("Accuratezza del modello:",accuracy*100,'%')

#For each y_test print the corresponding y_pred
for x, y in zip(y_test, y_pred):
    print(f"Elemento del test: {x}, Elemento della previsione: {y}")
```
